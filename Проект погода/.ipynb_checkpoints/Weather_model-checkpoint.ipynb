```python
 def days_from_data(data: str) -> int:
    answer = 0
    year, month, day = [int(i) for i in data.split('-')]
    answer += (year - START_DATA[0]) * 365
    for i in range(START_DATA[0], year+1):
        answer += i % 4 == 0 and (i % 100 != 0 or i % 400 == 0)
    for i in range(START_DATA[1], month):
        answer += DAY_IN_MONTH[i]
    answer += day - START_DATA[2]
    return answer
```

```python
data_len = len(data)
START_DATA = [1900, 1, 1]
DAY_IN_MONTH = {1: 31, 2: 28, 3: 31,
                4: 30, 5: 31, 6: 30,
                7: 31, 8: 31, 9: 30,
                10: 31, 11: 30, 12: 31}
```

```python
data = data.reset_index(drop=True)
```

```python
data['Date'] = data['Date'].apply(lambda x: days_from_data(x))
```

```python
max_date = data["Date"].max()
```

```python
data['Date'] = data['Date'] / max_date
```
